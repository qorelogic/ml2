```python
import pandas as p
import numpy as n

def generatePortfolioBreakdown():
    # obtainable from: https://openbook.etoro.com/manapana/portfolio/
    portfolioBreakdown = []
    # copied from Portfolio breakdown numbers (copy the first two)
    cd = p.DataFrame(n.array(p.read_csv('/ml/bin/etoro-equities.log', header=None)))
    plot(cd)
    c = cd.ix[len(cd)-1,:].tolist()
    profit = abs(c[0]-c[1])
    c.append(profit)
    c.append(profit/c[0]*100)
    portfolioBreakdown.append(c)
    portfolioBreakdown.append([0, 50.93, 0, 0])
    
    # Other (padding for the time being)
    portfolioBreakdown.insert(1, [0, 0, 0, 0])
    portfolioBreakdown.insert(len(portfolioBreakdown), sum(n.array(portfolioBreakdown),0).tolist())
    
    portfolioBreakdown = p.DataFrame(portfolioBreakdown, index=['noasnoas', 'Markets', 'Available Balance', 'Total'], columns=['TotalAllocated', 'Equity', 'Current Profit', 'Profit %'])
    equity             = p.DataFrame(list(portfolioBreakdown.ix['Total',['Equity','Current Profit','Profit %']]), index=['equity', 'equityUp', 'equityUpPercent'], columns=[0])
    dAllocation        = n.array(portfolioBreakdown.ix[0:3,['Equity']].transpose())[0].tolist()
    
    dAllocation.insert(2,0)
    dAllocation = dAllocation / sum(dAllocation) * 100
    
    print
    print 'Equity'
    print equity
    print
    #print 'Allocation (%)'
    #print p.DataFrame(dAllocation, index=['People', 'Stocks', 'Others', 'Balance'], columns=['-'])
    print 'Allocation (summary)'
    print p.DataFrame(n.array(portfolioBreakdown.ix[0:3,['Equity','Current Profit', 'Profit %']].transpose()), index=['amount', 'up', 'upPercent'], columns=['people', 'markets', 'balance'])
    print
    print 'Portfolio breakdown'
    print portfolioBreakdown
    print

generatePortfolioBreakdown()
```
