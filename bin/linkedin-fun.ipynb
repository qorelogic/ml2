```python
import pandas as p
import numpy as n
import popen2 as pop
from pylab import rcParams
rcParams['figure.figsize'] = 20, 10

def shingleString(s,k=2, increment=False):
    # k: the set of strings of length k
    d = {}
    li = []
    for i in range(0,len(s)-(k-1)):
        lo = s[i:i+k]
        li.append(lo)
        try:
            if increment:
                # if the increment is set to True do not increment
                raise
            d.update({lo:d[lo]+1})
        except:
            d.update({lo:1})
    #print d
    #print li
    #print dir(d.keys())
    
    #print d.keys()
    #print d.values()
    return d

def loadProfile(fname, label):
    t = open(fname).read()
    sh = shingleString(t,3)
    #ids = sh.keys()
    r = p.DataFrame(sh.values(), index=sh.keys(), columns=[label])
    plot(r,'.')
    return r

hdir = '/coursera/linked-in-profiles/'
pp = p.DataFrame()

d = pop.popen2('find '+hdir+' -type f')[0].read().split('\n')
d.pop()
cnt = 0
for i in d:
    cnt += 1
    #print i
    pp = pp.combine_first(loadProfile(i, cnt))

pp = pp.fillna(0).transpose()
pp
```
