```python
from pylab import rcParams
rcParams['figure.figsize'] = 20, 10

def shingleString(s,k=2, increment=False):
    # k: the set of strings of length k
    d = {}
    li = []
    for i in range(0,len(s)-(k-1)):
        lo = s[i:i+k]
        li.append(lo)
        try:
            if increment:
                # if the increment is set to True do not increment
                raise
            d.update({lo:d[lo]+1})
        except:
            d.update({lo:1})
    #print d
    #print li
    #print dir(d.keys())
    
    #print d.keys()
    #print d.values()
    return d

import pandas as p
import numpy as n

def loadProfile(fname, label):
    t = open(fname).read()
    sh = shingleString(t,3)
    #ids = sh.keys()
    r = p.DataFrame(sh.values(), index=sh.keys(), columns=[label])
    plot(r,'.')
    return r

pp = p.DataFrame()

import popen2 as pop
hdir = '/coursera/linked-in-profiles/ml-007/'
d = pop.popen2('ls '+hdir)[0].read().split('\n')
d.pop()
cnt = 0
for i in d:
    cnt += 1
    f = hdir+i
    print f
    pp = pp.combine_first(loadProfile(f, cnt))

"""
pp = pp.combine_first(loadProfile('/coursera/linked-in-profiles/ml-007/Alexei Vasilkov   LinkedIn.html', 1))
pp = pp.combine_first(loadProfile('/coursera/linked-in-profiles/ml-007/Anthony Proschka   LinkedIn.html', 2))
pp = pp.combine_first(loadProfile('/coursera/linked-in-profiles/ml-007/David López Companys   LinkedIn.html', 3))
pp = pp.combine_first(loadProfile('/coursera/linked-in-profiles/ml-007/Guillaume Lebret   LinkedIn.html', 4))
"""
pp = pp.fillna(0).transpose()
pp
```
